## Define inputs

In this example we are merging four .tif tiles with matching .tfw georeferencing worldfiles.

```python
im1 = PIL.Image.open('../../resources/merge-geotiffs/ras250_gb/data/SR.tif')
im2 = PIL.Image.open('../../resources/merge-geotiffs/ras250_gb/data/SS.tif')
im3 = PIL.Image.open('../../resources/merge-geotiffs/ras250_gb/data/SW.tif')
im4 = PIL.Image.open('../../resources/merge-geotiffs/ras250_gb/data/SX.tif')
```
